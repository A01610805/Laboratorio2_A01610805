```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('comprar_alquilar.csv')
df.info()
```

```python
ingresos = df['ingresos']
gc = df['gastos_comunes']
pc = df['pago_coche']
go = df['gastos_otros']
ahorros = df['ahorros']
vivienda = df['vivienda']
ec = df['estado_civil']
hijos = df['hijos']
trabajo = df['trabajo']
comprar = df['comprar']
```

```python
y1 = np.mean(ingresos)
des1 = np.std(ingresos)
mx1 = np.max(ingresos)
mn1 = np.min(ingresos)

y2 = np.mean(gc)
des2 = np.std(gc)
mx2 = np.max(gc)
mn2 = np.min(gc)

y3 = np.mean(pc)
des3 = np.std(pc)
mx3 = np.max(pc)
mn3 = np.min(pc)

y4 = np.mean(go)
des4 = np.std(go)
mx4 = np.max(go)
mn4 = np.min(go)

y5 = np.mean(ahorros)
des5 = np.std(ahorros)
mx5 = np.max(ahorros)
mn5 = np.min(ahorros)

y6 = np.mean(vivienda)
des6 = np.std(vivienda)
mx6 = np.max(vivienda)
mn6 = np.min(vivienda)

y7 = np.mean(ec)
des7 = np.std(ec)
mx7 = np.max(ec)
mn7 = np.min(ec)

y8 = np.mean(hijos)
des8 = np.std(hijos)
mx8 = np.max(hijos)
mn8 = np.min(hijos)

y9 = np.mean(trabajo)
des9 = np.std(trabajo)
mx9 = np.max(trabajo)
mn9 = np.min(trabajo)

y10 = np.mean(comprar)
des10 = np.std(comprar)
mx10 = np.max(comprar)
mn10 = np.min(comprar)
```
